Importing the necessary libraries

```python
import pandas as pd
import numpy as np
from datetime import datetime
from google.colab import files
```

Loading the dataset

```python
df = pd.read_csv('train.csv')
df
```

Detect the Missing Values

```python
missing_summary = df.isnull().sum()
print(missing_summary)
```

a) Look for the missing values in all the columns and either impute them (replace with mean, median, or mode) or drop them. Justify your action for this task.

```python
for column in df.columns:
    if df[column].isna().sum() > 0:
        if df[column].dtype == 'object':
            mode_value = df[column].mode()[0]
            df[column] = df[column].fillna(mode_value)
            print(f"• {column}: filled missing values with MODE = '{mode_value}'")
        else:
            mean_value = df[column].mean()
            df[column] = df[column].fillna(mean_value)
            print(f"• {column}: filled missing values with MEAN = {round(mean_value, 2)}")

# Check missing values after
print(df.isna().sum())
```

b) Remove the units from some of the attributes and only keep the numerical values (for example remove kmpl from “Mileage”, CC from “Engine”, bhp from “Power”, and lakh from “New_price”).

```python
# Remove units and convert to numeric values
df['Mileage'] = pd.to_numeric(df['Mileage'].str.replace(' kmpl', '', regex=False).str.replace(' km/kg', '', regex=False), errors='coerce')

df['Engine'] = pd.to_numeric(df['Engine'].str.replace(' CC', '', regex=False), errors='coerce')

df['Power'] = pd.to_numeric(df['Power'].str.replace(' bhp', '', regex=False), errors='coerce')

df['New_Price'] = pd.to_numeric(df['New_Price'].str.replace(' lakh', '', regex=False), errors='coerce')
```

C) Change the categorical variables (“Fuel_Type” and “Transmission”) into numerical one hot encoded value.

```python

# Encode Diesel Fuel_Type as 1, Petrol Fuel_Type as 0
df.loc[df.Fuel_Type=='Diesel','Fuel_Type']=1
df.loc[df.Fuel_Type=='Petrol','Fuel_Type']=0
# Encode Manual Transmission as 1, Automatic Transmission as 0'
df.loc[df.Transmission=='Manual','Transmission']=1
df.loc[df.Transmission=='Automatic','Transmission']=0
```

Create a new feature 'Car_Age' by subtracting 'Year' from the current year

```python
current_year = datetime.now().year

df['Car_Age'] = current_year - df['Year']
df
```

Perform select, filter, rename, mutate, arrange and summarize with group by operations (or
their equivalent operations in python) on this dataset.

```python
# 1. Select specific columns
selected_columns = df[['Name', 'Fuel_Type', 'Price']]
print("Selected Columns:\n", selected_columns.head())  # Display first 5 rows for brevity

# 2. Filter rows where Fuel_Type is Diesel
filtered_df = df[df['Fuel_Type'] == 'Diesel']
print("\nFiltered DataFrame (Diesel Cars):\n", filtered_df.head())

# 3. Rename columns
renamed_df = df.rename(columns={'Kilometers_Driven': 'Distance_Driven', 'Owner_Type': 'Owner'})
print("\nRenamed DataFrame:\n", renamed_df.head())

# 4. Mutate: Add a new column for Mileage in float (remove 'kmpl' or 'km/kg')
df['Mileage_float'] = pd.to_numeric(df['Mileage'].astype(str).str.replace(r' kmpl| km/kg', '', regex=True), errors='coerce')
print("\nDataFrame after Mutate (New Mileage Column):\n", df[['Mileage', 'Mileage_float']].head())

# 5. Arrange: Sort by Price in descending order
arranged_df = df.sort_values(by='Price', ascending=False)
print("\nArranged DataFrame (Sorted by Price):\n", arranged_df[['Name', 'Price']].head())

# 6. Summarize with Group By: Average Price by Fuel_Type
summary_df = df.groupby('Fuel_Type')['Price'].mean().reset_index()
print("\nAverage Price by Fuel Type:\n", summary_df)
```
